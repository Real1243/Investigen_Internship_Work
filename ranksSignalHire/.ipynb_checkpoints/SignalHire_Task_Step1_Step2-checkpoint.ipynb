```python
import requests
import json


API_KEY = "202.M3D7xK6FykhOIHuKoFGLLjMI8jwW"  


BASE_URL = "https://www.signalhire.com/api/v1"
SEARCH_ENDPOINT = "/candidate/searchByQuery"  #for the Search API
PERSON_ENDPOINT = "/candidate/search"  # for Person API

def get_seniority_rank(title):
    if not title:
        return 999
    title_lower = title.lower()
    if any(w in title_lower for w in ['chief', 'cfo', 'chief financial', 'chief treasury']):
        return 1
    if any(w in title_lower for w in ['vp', 'vice president', 'svp', 'evp']):
        return 2
    if any(w in title_lower for w in ['director', 'head of', 'group head']):
        return 3
    if any(w in title_lower for w in ['senior manager', 'lead', 'principal']):
        return 4
    if any(w in title_lower for w in ['manager']):
        return 5
    return 6

def search_finance_treasury(company_name, size=10):
    headers = {"apikey": API_KEY, "Content-Type": "application/json"}
    payload = {
        "keywords": "finance OR treasury",
        "currentTitle": "CFO OR \"Vice President Finance\" OR \"Finance Director\" OR \"Treasury Director\" OR \"Group Treasurer\" OR Manager OR Director",
        "currentCompany": f"\"{company_name}\"",
        "size": size,
        
    }
    response = requests.post(BASE_URL + SEARCH_ENDPOINT, headers=headers, json=payload)
    if response.status_code != 200:
        print("Search error:", response.text)
        return []
    data = response.json()
    profiles = data.get("profiles", [])
    
    def get_current_title(p):
        exp = p.get("experience", [])
        return exp[0].get("title", "") if exp else ""
    
    sorted_profiles = sorted(profiles, key=lambda p: get_seniority_rank(get_current_title(p)))
    return sorted_profiles[:10]  # only top 10 for now
```

```python

CALLBACK_URL = "https://uninwrapped-celinda-subcranially.ngrok-free.dev"  # REPLACE!

def reveal_contacts(linkedin_urls):
    if not linkedin_urls:
        print("No LinkedIn URLs to reveal.")
        return
    headers = {"apikey": API_KEY, "Content-Type": "application/json"}
    payload = {
        "items": linkedin_urls,  # list of LinkedIn URLs from profiles
        "callbackUrl": CALLBACK_URL,
        # "withoutContacts": false  # default is with contacts
    }
    response = requests.post(BASE_URL + PERSON_ENDPOINT, headers=headers, json=payload)
    print("Reveal request status:", response.status_code)
    print("Response:", response.text)
    print("Wait for callback to your webhook URL - check there for full data (emails/phones).")
    print("Once received, parse JSON from callback for contacts.")

# Main interactive flow

# Main flow - top 10 list + manual selection
company = input("Enter company name (e.g., HDFC Bank): ")
top_profiles = search_finance_treasury(company, size=20)

if not top_profiles:
    print("No results found.")
else:
    print(f"\nFound {len(top_profiles)} profiles for {company}")
    print("Top 10 most senior (sorted by title seniority):\n")
    
    # Store UIDs and basic info
    person_map = {}
    for idx, p in enumerate(top_profiles[:10], 1):
        name = p.get("fullName", "N/A")
        title = p.get("experience", [{}])[0].get("title", "N/A")
        uid = p.get("uid")
        
        print(f"{idx}. {name} - {title}")
        if uid:
            person_map[idx] = {"uid": uid, "name": name, "title": title}
        else:
            print("   (No UID - cannot reveal contacts)")
        print()
    
    # Reveal all 10 at once (if UID exists)
    reveal_uids = [info["uid"] for info in person_map.values() if info["uid"]]
    if reveal_uids:
        print("\nRevealing contacts for all 10 people (async - wait for Flask terminal)...")
        reveal_contacts(reveal_uids)
    else:
        print("No UIDs found - no contacts can be revealed.")
    
    print("\nTo see details of a person:")
    print("Type the number (1-10) and press Enter")
    print("Details will appear in Flask terminal (callback_server.py) as they arrive")
    print("Type 'exit' to quit\n")
    
    while True:
        user_input = input("Enter person number (1-10) or 'exit': ").strip()
        if user_input.lower() == 'exit':
            print("Exiting.")
            break
        if not user_input.isdigit():
            print("Please enter a number 1-10 or 'exit'")
            continue
        num = int(user_input)
        if num < 1 or num > 10 or num not in person_map:
            print("Invalid number - choose 1-10")
            continue
        
        person = person_map[num]
        print(f"\nYou selected: {num}. {person['name']} - {person['title']}")
        print("Waiting for callback details in Flask terminal...")
        print("(If already arrived, check the Flask terminal now)")
        print("Details will show: Full Name + Current Experience + Contacts (work emails + phones)\n")
```
